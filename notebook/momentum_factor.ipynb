```python
import sys
import pandas as pd
from pathlib import Path

sys.path.append('..')
from src.backtest import MultiFactorBacktest

data_dir = Path('../data')
price_data = {}

for csv_file in data_dir.glob('*.csv'):
    code = csv_file.stem
    df = pd.read_csv(csv_file)
    
    # 改成你实际的列名：trade_date
    df['date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
    df = df.sort_values('date')
    
    price_data[code] = df

print(f"成功加载了 {len(price_data)} 只股票的数据，例如：{list(price_data.keys())[:3]}")

bt = MultiFactorBacktest(
    price_data=price_data,
    start_date='2023-01-01',
    end_date='2024-12-31',
    top_n=2
)

result = bt.run_backtest()

print(f"策略总收益率: {result['total_return']:.2%}")
print(f"持有天数: {len(result['daily_returns'])}")
```
